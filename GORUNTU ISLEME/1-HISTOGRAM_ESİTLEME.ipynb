Histogram Eşitleme:Görüntüdeki renk dağılımını düzenleyerek kontrast dengesini iyileştirmede histogram eşitleme yöntemi kullanılmakta.

Histogram eşitleme, local ve global olmak üzere iki fonksiyona ayrılmaktadır.

CLAHE histogram eşitleme yöntemi kulllanılarak örnek bir görüntü üzerinde histogram eşitleme işlemi gerçekleştirilip

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
goruntu = cv2.imread('/content/Ak (50).png',0)
clahe =cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
equ =clahe.apply(goruntu)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(goruntu)
cv2_imshow(equ)
```

```python
ax1=plt.subplot(221)
ax1.set_title("Orijinal Goruntu Histogramı",fontsize=10)
plt.hist(goruntu.flatten(),256,[0,256],color='r')
```

```python
ax2 = plt.subplot(222)
ax2.set_title("Eşitlenmiş Görüntü Histogramı",fontsize=10)
ax2.hist(equ.flatten(),256,[0,256],color='b')
plt.show()
```
