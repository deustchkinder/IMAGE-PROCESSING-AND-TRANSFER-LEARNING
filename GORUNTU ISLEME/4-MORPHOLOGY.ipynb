MORFOLOJİ:Görüntüdeki nesnelerin geometrik şekline dayalı kümeler teorisi ile oluşturulmuş önemli görüntü işleme yöntemidir.

Morfolojik süzgeçleme görüntü işlemede,inceltme,erozyon gibi amaçlarla ön ya da son işlem olarak kullanılır. 

Siyah-Beyaz zaten vardır ek olarak Gri Seviyeli Görüntüler üzerinde de kullanılabilir.

```python
import cv2
from pylab import *
img = cv2.imread('/content/Ak (50).png',0)
subplot(331)
imshow(img,cmap="gray")
xticks([])
yticks([])
kernel = ones((5,5),uint8)
```

Aşındırma işlemi, ikili görüntülerdeki düşük değerler çıkartılarak nesnenin küçülmesini veya incelmesini sağlayan morfolojik işlemlerdir.

Aşındırma işlemi ile görüntüdeki düşük değerler çıkartılarak nesnenin küçülmesini veya incelmesini sağlayan morfolojik işlem.

```python
erosion = cv2.erode(img,kernel,iterations = 1)
subplot(332)
imshow(erosion,cmap="gray")
xticks([])
yticks([])
```

Genişletme İşlemi, yüksek değerler eklenerek ikili görüntüdeki nesnenin büyümesini veya kalınlaşmasını sağlayan bir morfolojik işlemlerdir.

Sayısal bir görüntüyü genişletmek, görüntüyü yapısal eleman ile kesiştiği bölümler kadar büyütme işlemidir.

```python
dilation = cv2.dilate(img,kernel,iterations=1)
subplot(333)
imshow(dilation,cmap="gray")
xticks([])
yticks([])
```

Bir görüntüye ilk olarak aşınma işlemi, sonrasında ise elde edilen görüntüyü genişletme işleminin uygulanması ile Açma İşlemi elde edilmektedir.

Açma işlemi bir görüntü üzerinde bulunan gürültüleri azaltmak için kullanılan bir yöntem.

Kapama işlemi ise bir görüntüye önce genişletme işlemi ardından da aşındırma işlemi yapılan bir yöntemdir.

```python
opening = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
subplot(334)
imshow(opening,cmap="gray")
xticks([])
yticks([])
```

```python
closing = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
subplot(335)
imshow(closing,cmap="gray")
xticks([])
yticks([])
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
subplot(336)
imshow(gradient,cmap="gray")
xticks([])
yticks([])
```

```python
tophat = cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernel)
subplot(337)
imshow(tophat,cmap="gray")
xticks([])
yticks([])
```

```python
blackhat =cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kernel)
subplot(338)
imshow(blackhat,cmap="gray")
xticks([])
yticks([])
```
