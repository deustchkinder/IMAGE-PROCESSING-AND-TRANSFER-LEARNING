```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def apply_canny_edge_detection(image, threshold1, threshold2):
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    denoised = cv2.GaussianBlur(gray, (3, 3), 0)
    edges = cv2.Canny(denoised, threshold1, threshold2)

    return edges
```

```python
image = cv2.imread('/content/Ak (50).png')
```

```python
edges = apply_canny_edge_detection(image, 50, 150)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Orijinal Görüntü'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Kenarlar'), plt.xticks([]), plt.yticks([])
plt.show()
```
