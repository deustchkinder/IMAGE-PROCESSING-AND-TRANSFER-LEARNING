```python
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, exposure
import cv2
```

```python
img = cv2.imread('/content/Ak (50).png')
```

```python
_,hog_img = hog(img,orientations=8,pixels_per_cell=(16,16),cells_per_block=(1,1),visualize=True,multichannel=True)
```

```python
img = exposure.rescale_intensity(hog_img,in_range=(0,10))
```

```python
plt.axis('off')
plt.imshow(img,cmap=plt.cm.gray)
plt.show()
```
