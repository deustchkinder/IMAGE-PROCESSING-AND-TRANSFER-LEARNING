```python
import cv2
import matplotlib.pyplot as plt

def resize_image(image_path, output_path, new_width, new_height):
   
    image = cv2.imread('/content/Ak (50).png')
    resized_image = cv2.resize(image, (new_width, new_height))
    
    cv2.imwrite(output_path, resized_image)
    resized_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
    
    plt.imshow(resized_image)
    plt.axis('off')
    plt.show()

input_image = 'giris_goruntusu.jpg'
output_image = 'cikis_goruntusu.jpg'

new_width = 512
new_height = 512

resize_image(input_image, output_image, new_width, new_height)
```
